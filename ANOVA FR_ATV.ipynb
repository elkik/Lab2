```python
import seaborn as sns

# sns.lmplot permet de tracer une régression univariée
# voici une régression saisissant la relation entre une 
# variable et une autre, par exemple le salaire et l'éducation

sns.lmplot(y='ML_FoldInd', x='ML_Area', data=df)
```

```python
sns.pairplot(df, vars=["ML_FoldInd",'ML_Area', 'ML_Thick', 'ML_Vol'], kind='reg')  
```

```python
print(ols(formula='ML_FoldInd ~ ML_Area + ML_Thick + ML_Vol',
                data=df).fit().summary())
```

```python
model_anova = ols(formula='ML_FoldInd ~ ML_Area + ML_Thick + ML_Vol * ML_FoldInd',
                data=df).fit()

print(model_anova.summary())
```

```python
model_anova.pvalues.loc['ML_Vol:ML_FoldInd']
```
